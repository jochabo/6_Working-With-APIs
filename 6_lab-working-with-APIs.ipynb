```python
import pandas as pd
import requests
from pandas.io.json import json_normalize
import numpy as np
```

# 1) Finance - Econdb	Global macroeconomic data - European Union Consuper Price Index

```python
url = 'https://www.econdb.com/api/series/CPIEU/?format=json'
EUCPI = requests.get(url).json()
display(EUCPI)
type(EUCPI)
```

```python
EUCPI_df = pd.DataFrame(EUCPI)
display(EUCPI_df.shape)
EUCPI_df.head()
```

```python
flattened_data = json_normalize(EUCPI)
display(flattened_data.shape)
flattened_data
```

I am interested in the columns data.values and data.dates.

```python
col_list = ['data.values','data.dates']
col_list
```

```python
flattened_data['data.values'][0]
flattened_data['data.dates'][0]
```

```python
final_dataframe = pd.DataFrame()

for column in col_list:
    data_in_list = flattened_data[column][0]
    list_in_df = pd.DataFrame(data_in_list, columns=[column])
    final_dataframe = pd.concat([list_in_df, final_dataframe], axis=1)
final_dataframe
```
